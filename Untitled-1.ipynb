# States that are ahead on the electrical car stations
in this code we will try to find out which states in the U.S have more electrical car charging stations compared to other states.


```python
import pandas as pd
import seaborn as sb
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2022/2022-03-01/stations.csv")
```

```python
data
```

```python
elec_station=data[(data["FUEL_TYPE_CODE"]=="ELEC") & (data["COUNTRY"]=="US")]
```

```python
states=elec_station["STATE"].value_counts()
```

```python
states
```

```python
states.to_frame(name="count")
```

```python
states.index.name="state"
```

```python
states.to_frame(name="values")
```
